```python
%pylab inline
import scipy
from scipy.interpolate import *
```

```python
#Runge function
def myfun(x):
    return 1/((1+50*(x-.5)**2))
```

```python
N = 1025
M = 10

x=linspace(0,1,N)
q=linspace(0,1,M)

plot(x,myfun(x))
```

```python
#piece wise linear interpolation
yq= myfun(q)
p =interp1d(q,yq, kind='cubic') ##'linear' 'quadratic' 'cubic'

plot(x,myfun(x))
plot(q,yq,'ro')
plot(x,p(x))
```

```python
B = zeros((N,M))
dB= zeros((N,M))

for i in range(M):
    c=q*0
    c[i]=1
    bi=PPoly.from_spline(splrep(q,c,k=2))
    B[:,i]=bi(x)
    dB[:,i]=bi.derivative(1)(x)
   # plot(x,bi.derivative(0)(x))
   # plot(q,c,'ro')
_ = plot(x,B)
```

```python
plot(x,B.dot(myfun(q)))
```

```python
BtB = B.T.dot(B)


pL2= linalg.solve(BtB,B.T.dot(myfun(x)))
```

```python
plot(x,myfun(x))
plot(x,B.dot(myfun(q)))
plot(x,B.dot(pL2))
plot(q,myfun(q),'ro')
plot(q,pL2,'go')
```

```python
A = dB.T.dot(dB)+B.T.dot(B)
```

```python
#the constant one
rhs = B.T.dot(x*0+1)#right hand side
u = linalg.solve(A,rhs)

plot(x,B.dot(u))
```

```python
rhs= B.T.dot(x)
u=linalg.solve(A,rhs)
plot(x,B.dot(u))
```
