```python
N_data_points = 1000

x= linspace(0,1,N_data_points)

y=sin(2*pi*x)*0.5
```

```python
plot(x,y)
```

```python
degree=4

basis = zeros((N_data_points,degree+1))

for d in range (degree+1):
    basis[:,d]=x**d
```

```python
plot(x,basis)
```

Solve the problem: find p in P^degree, such taht p is at "minimum distance from data points"

p= sum_{i=0}^degree basis_i*p_i

basis_i=x^i

```python
def Loss(p,y):
    return .5*(basis.dot(p)-y).dot(basis.dot(p)-y)
```

```python
BtB= basis.T.dot(basis)
Bty=basis.T.dot(y)
```

```python
One = basis[:,0]
```

```python
OneTOne=One.T.dot(One)
```

```python
p= One.T.dot(y)/OneTOne
```

```python
p= linalg.solve(BtB,Bty)
```
